```python
for _ in range(100):
     
    f = FBM(n=1024, hurst=0.75, length=1, method='daviesharte')
    # Generate a fBm realization
    fbm_sample = f.fbm()

    # Generate a fGn realization
    fgn_sample = f.fgn()

    # Get the times associated with the fBm
    t_values = f.times()
    xo = 8000
    mu = 0.0073
    sigma = 0.0938
    hurst_sigma = 0.0860
    hurst = 0.5713 + np.random.normal(0.0, hurst_sigma, 1)[0]
    prices = []
    for i in range(len(fbm_sample)):
        H = fbm_sample[i]
        t = t_values[i]

        xt = xo * math.exp( mu * t + sigma * H)
        prices.append(xt)

    plt.title('FBM')
    plt.xlabel('time (days)', fontsize=14)
    plt.ylabel('Price', fontsize=14)
    plt.plot(t_values, prices)
plt.show()
```
